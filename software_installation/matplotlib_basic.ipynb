# Matplotlib Basic

[Matplotlib](http://matplotlib.org/) is an excellent 2D and 3D graphics library for generating scientific figures.

```python
# This line configures matplotlib to show figures embedded in the notebook, instead of opening a new window for each figure.
%matplotlib inline
```

## 2D Figures

```python
from pylab import *
import numpy as np
I = np.identity(3)
```

```python
x = linspace(0, 5, 10)
y = x ** 2
```

```python
figure()
plot(x, y, '--r')
xlabel('x')
ylabel('y')
title('title')
show()
```

```python
figure()
x = np.random.random(50)
y = x * 10 + np.random.random(50)
scatter(x, y)
show()
```

```python
figure()
hist(np.random.randn(10000))
show()
```

## 3D Figures

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
x = linspace(0, 5, 100)
y = x ** 2
z = np.sin(x)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
p = ax.plot(x, y, z)
```
